**SINGLE DATE HISTORICAL DATA**

```python
import requests
import csv
from bs4 import BeautifulSoup
from datetime import datetime
url = 'https://dps.psx.com.pk/historical'

current_date =  '2021-08-13'

data = {
    'date' : current_date
}

response = requests.post(url, data = data)
# print(response)
# print(response.text)

soup = BeautifulSoup(response.text, 'html.parser')

rows = soup.findAll('tr')

main_data = []

for row in rows[1:]:
    row_data = []
    for items in row.findAll('td'):
        row_data.append(items.text)
    row_data.append(current_date)
    row_data.append(datetime.now().strftime('%d/%m/%Y %H:%M:%S'))    
    main_data.append(row_data)

# print(main_data[:6])
# print(len(main_data))

file = open(r'C:\Users\wwwmu\Downloads\singledate_historical.csv' , 'a+', newline='')
with file:
    write = csv.writer(file)
    write.writerows(main_data)
```

**MONTH WISE YEARLY HISTORICAL DATA BASED ON SYMBOL**

```python
import requests
import csv
from bs4 import BeautifulSoup
from datetime import datetime

url = 'https://dps.psx.com.pk/historical'
symbol = 'TELE'

start_year = 2024
start_month = 4
end_year = 2025
end_month = 3

file = open(r'C:\Users\wwwmu\OneDrive\Desktop\Data from Symbol.csv', 'w', newline='', encoding='utf-8')
with file:
    write = csv.writer(file)
    write.writerow(['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Symbol', 'Scraped At'])

    for year in range(start_year, end_year + 1):
        for month in range(1, 13):
            if (year == start_year and month < start_month) or (year == end_year and month > end_month):
                continue

            print(f"Scraping {symbol} for {month:02d}/{year}")
            data = {
                'month': month,
                'year': year,
                'symbol': symbol
            }

            response = requests.post(url, data=data)
            soup = BeautifulSoup(response.text, 'html.parser')
            rows = soup.findAll('tr')

            for row in rows[1:]:
                row_data = []
                for items in row.findAll('td'):
                    row_data.append(items.text.strip())
                row_data.append(symbol)
                row_data.append(datetime.now().strftime('%d/%m/%Y %H:%M:%S'))
                write.writerow(row_data)

print("Scraping completed successfully.")
```
