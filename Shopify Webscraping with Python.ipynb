```python
import requests
import json
import pandas as pd
```

```python
url = 'https://helmboots.com/products.json?limit=200&page=1'
r = requests.get(url)
data = r.json()
print(data)
```

```python
# Extracting the first title

print(data['products'][0]['title'])
```

```python
# Extracting all the titles

for item in data['products']:
        print(item['title'])
```

```python
for item in data['products']:
        title = item['title']
        handle = item['handle']
        created = item['created_at']
        product_type = item['product_type']
        print(title, handle, created, product_type)
    
```

```python
product_list = []  # Creating a list
for item in data['products']:
        title = item['title']
        handle = item['handle']
        created = item['created_at']
        product_type = item['product_type']
        # print(title, handle, created, product_type)
        for image in item['images']:
            try:
                image_src = image['src']
            except:
                image_src = 'None'
        for variant in item['variants']:
            price = variant['price']
            sku = variant['sku']
            available = variant['available']
            product = {
                    "title" : title,
                    "handle" : handle,
                    "created" : created,
                    "product_type" : product_type,
                    "image" : image_src,
                    "price" : price,
                    "sku" : sku,
                    "available" : available
                        }
            
            product_list.append(product)
            
print(product_list)
```

```python
# Creating a df and then converting it to a CSV File

df = pd.DataFrame(product_list)
df.to_csv('data_shopifywebscraping.csv')
```
